```python
import zipfile
import os

zip_file = 'archive.zip'
extract_path = '/content/extracted_files'
with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

os.listdir(extract_path)
```

```python
import pandas as pd
csv_file_path = os.path.join(extract_path, 'Global_Education.csv')
education_data = pd.read_csv(csv_file_path, encoding='latin1')
education_data.head()
```

```python
education_data.info()
education_data.isnull().sum()
education_data.describe()
```

```python
education_data = education_data.drop_duplicates()
```

```python
numerical_columns = education_data.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = education_data.select_dtypes(include=['object']).columns
education_data[numerical_columns] = education_data[numerical_columns].fillna(education_data[numerical_columns].mean())
education_data[categorical_columns] = education_data[categorical_columns].fillna('Unknown')


```

```python
numerical_columns = education_data.select_dtypes(include=['float64', 'int64']).columns
Q1 = education_data[numerical_columns].quantile(0.25)
Q3 = education_data[numerical_columns].quantile(0.75)
IQR = Q3 - Q1
education_data = education_data[~((education_data[numerical_columns] < (Q1 - 1.5 * IQR)) |
                                  (education_data[numerical_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

```

```python
from sklearn.preprocessing import StandardScaler
numerical_columns = education_data.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
education_data[numerical_columns] = scaler.fit_transform(education_data[numerical_columns])

```

```python
education_data = pd.get_dummies(education_data, columns=categorical_columns, drop_first=True)

```

```python
education_data.to_csv('processed_education_data.csv', index=False)
from google.colab import files
files.download('processed_education_data.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns
numerical_columns = education_data.select_dtypes(include=['float64', 'int64']).columns

# Create box plots for numerical columns
plt.figure(figsize=(20, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, (len(numerical_columns) // 3) + 1, i)  # Dynamically adjust grid size
    sns.boxplot(data=education_data, y=col)
    plt.title(f'Box Plot - {col}', fontsize=12)
    plt.tight_layout()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting only the relevant columns for the correlation
oosr_columns = [
    'OOSR_Pre0Primary_Age_Male', 'OOSR_Pre0Primary_Age_Female',
    'OOSR_Primary_Age_Male', 'OOSR_Primary_Age_Female',
    'OOSR_Lower_Secondary_Age_Male', 'OOSR_Lower_Secondary_Age_Female',
    'OOSR_Upper_Secondary_Age_Male', 'OOSR_Upper_Secondary_Age_Female'
]

# Calculate correlation matrix
corr = education_data[oosr_columns].corr()

# Create a heatmap to visualize the correlation
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Between Out-of-School Rates for Different Education Stages')
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reshaping the data for better plotting
melted_data = education_data.melt(
    id_vars=[],
    value_vars=[
        'OOSR_Pre0Primary_Age_Male', 'OOSR_Pre0Primary_Age_Female',
        'OOSR_Primary_Age_Male', 'OOSR_Primary_Age_Female',
        'OOSR_Lower_Secondary_Age_Male', 'OOSR_Lower_Secondary_Age_Female',
        'OOSR_Upper_Secondary_Age_Male', 'OOSR_Upper_Secondary_Age_Female'
    ],
    var_name='Category',
    value_name='Out_of_School_Rate'
)

# Add new columns for Education Stage and Gender by extracting from 'Category'
melted_data['Education_Stage'] = melted_data['Category'].apply(lambda x: ' '.join(x.split('_')[1:3]))
melted_data['Gender'] = melted_data['Category'].apply(lambda x: x.split('_')[3])

# Check if the reshaped data looks correct
print(melted_data.head())

# Plotting the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Education_Stage', y='Out_of_School_Rate', hue='Gender', data=melted_data, palette='Set2')
plt.title('Out-of-School Rates by Education Stage and Gender')
plt.xticks(rotation=45)
plt.show()


```

```python
import plotly.graph_objects as go
x=[1,2,3,4,5]
y=[2,3,5,7,11]

fig=go.Figure(data=go.Scatter(x=x,y=y,mode='lines+markers'))

fig.update_layout(
    title='Simple line plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)
fig.show()
```

```python
# prompt: get me bar chart using plotly for students and marks example

import plotly.graph_objects as go

students = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
marks = [85, 92, 78, 65, 95]

fig = go.Figure(data=[go.Bar(x=students, y=marks)])

fig.update_layout(
    title='Student Marks',
    xaxis_title='Students',
    yaxis_title='Marks'
)

fig.show()
```

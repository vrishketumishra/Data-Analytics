```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/EdStatsCountry-Series.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, silhouette_score
```

```python
le = LabelEncoder()
df['CountryCode'] = le.fit_transform(df['CountryCode'])
df['SeriesCode'] = le.fit_transform(df['SeriesCode'])
```

```python
df['DESCRIPTION'] = le.fit_transform(df['DESCRIPTION'])
```

```python
X = df[['CountryCode', 'DESCRIPTION']]
y = df['SeriesCode']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Algorithm 1: KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("KNN Accuracy:", accuracy)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust clusters as needed
kmeans.fit(X)
silhouette = silhouette_score(X, kmeans.labels_)
print("K-Means Silhouette Score:", silhouette)
```
